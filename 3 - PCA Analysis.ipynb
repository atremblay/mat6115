```python
import pickle
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use('ggplot')
```

```python
PATH_1_LAYER = Path('lstm_1layer/')
PATH_2_LAYER = Path('lstm_2layer/')
```

```python
with open(PATH_1_LAYER / 'pca_vanilla.pkl', 'rb') as pca_vanilla_file:
    pca_vanilla_1_layer = pickle.load(pca_vanilla_file)

with open(PATH_1_LAYER / 'pca_trained.pkl', 'rb') as pca_trained_file:
    pca_trained_1_layer = pickle.load(pca_trained_file)
```

```python
with open(PATH_2_LAYER / 'pca_vanilla.pkl', 'rb') as pca_vanilla_file:
    pca_vanilla_2_layer = pickle.load(pca_vanilla_file)

with open(PATH_2_LAYER / 'pca_trained.pkl', 'rb') as pca_trained_file:
    pca_trained_2_layer = pickle.load(pca_trained_file)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5), sharex=True)
x = range(pca_trained_1_layer[0].explained_variance_ratio_.shape[0])
ax[0].scatter(
    x, pca_trained_2_layer[0].explained_variance_ratio_.cumsum(), s=5
)

ax[0].scatter(
    x, pca_trained_1_layer[0].explained_variance_ratio_.cumsum(), s=5
)
ax[0].legend(['2 layer LSTM', '1 layer LSTM'])
ax[0].set_title('First layer')
ax[0].set_ylabel('Explained Variance')
ax[0].set_xlabel('PCA Components')

ax[1].scatter(
    x, pca_trained_2_layer[1].explained_variance_ratio_.cumsum(), s=5
)
ax[1].legend(['2 layer LSTM'])
ax[1].set_title('Second layer')
ax[1].set_xlim((-1, 50))
ax[1].set_xlabel('PCA Components')
```
